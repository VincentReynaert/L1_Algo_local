```python
import pandas as pd
```

```python
df = pd.read_csv("notesL1A.txt",sep=",",index_col=["nom","prenom"])
df.index = df.index.map(lambda x:(x[0].lower(),x[1].lower()))
df = df.sort_index()
```

```python
df["qcm"] = df["qcm"].map(lambda x: eval(x))
df["exo"] = df["exo"].map(lambda x: eval(x))
df["notes"] = (df["qcm"] + df["exo"]).map(lambda x: min(x,20))
df
```

```python
df.to_csv("notesDMFinalL1A.csv")
```
