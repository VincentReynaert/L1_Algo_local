```python
import pandas as pd
```

```python
df = pd.read_csv("GroupeA/notesL1A.txt",sep=",",index_col=["nom","prenom"])
df.index = df.index.map(lambda x:(x[0].lower(),x[1].lower()))
df = df.sort_index()
df["qcm"] = df["qcm"].map(lambda x: eval(x))
df["exo"] = df["exo"].map(lambda x: eval(x))
df["notes"] = (df["qcm"] + df["exo"] ).map(lambda x: min(x if x==0  else x+0.5,20))
df.to_csv("notesDMFinalL1A.csv")
print(round(df["notes"].mean(),2),round(df["notes"].std(),2))
df
```
